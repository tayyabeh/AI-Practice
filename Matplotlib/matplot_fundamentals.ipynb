```python
# plt.scatter on pandas data 
bat = pd.read_csv(r'M:\tutorial\Machine learining\Matplotlib\Datasets\batter.csv')
b = bat.head(50)
plt.scatter(b['avg'],b['strike_rate'],color = 'green')
plt.title('Avg vs Strike Rate of top 50 players')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')
```

```python
# size 
tips = sns.load_dataset('tips')
plt.scatter(tips['total_bill'],tips['tip'],s = tips['size']*10, color = 'purple')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

Bar chart 
- Bivariate Analysis
- Numerical vs Categorical 
- Use Case - Aggregate analysis of groups

```python
# simple bar chart 
children = [10, 20, 30, 40, 50]
colors = ['red', 'blue', 'green', 'yellow', 'purple']
plt.bar(range(len(children)), children, color=colors)
plt.title('Children')
plt.xlabel('Children')
plt.ylabel('Count')
```

```python
# horizontl barcahrt ( increased category )
plt.barh(range(len(children)), children, color=colors)
```

```python
bat_season_record = pd.read_csv(r'M:\tutorial\Machine learining\Matplotlib\Datasets\batsman_season_record.csv')
bat_season_record
plt.bar(np.arange(bat_season_record.shape[0])- 0.2, bat_season_record['2015'],width=0.2)
plt.bar(np.arange(bat_season_record.shape[0]), bat_season_record['2016'],width=0.2)
plt.bar(np.arange(bat_season_record.shape[0])+ 0.2, bat_season_record['2017'],width=0.2)
plt.xticks(np.arange(bat_season_record.shape[0]), bat_season_record['batsman'])


plt.show()
```

```python
# problem 
# simple bar chart 
children = [10, 20, 30, 40, 50]
colors = ['red red red red red', 'blue blue blue blue blue', 'green green green green green', 'yellow yellow yellow yellow yellow', 'purple purple purple purple purple']
plt.bar(colors, children, color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked bar chart 
plt.bar(bat_season_record['batsman'], bat_season_record['2017'],label='2017')
plt.bar(bat_season_record['batsman'], bat_season_record['2016'],bottom=bat_season_record['2017'],label='2016')
plt.bar(bat_season_record['batsman'], bat_season_record['2015'],bottom=bat_season_record['2016']+bat_season_record['2017'],label='2015')
plt.legend()
```
