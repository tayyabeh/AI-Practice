Bar chart 
- Bivariate Analysis
- Numerical vs Categorical 
- Use Case - Aggregate analysis of groups

```python
# simple bar chart 
children = [10, 20, 30, 40, 50]
colors = ['red', 'blue', 'green', 'yellow', 'purple']
plt.bar(range(len(children)), children, color=colors)
plt.title('Children')
plt.xlabel('Children')
plt.ylabel('Count')
```

```python
# horizontl barcahrt ( increased category )
plt.barh(range(len(children)), children, color=colors)
```

```python
bat_season_record = pd.read_csv(r'M:\tutorial\Machine learining\Matplotlib\Datasets\batsman_season_record.csv')
bat_season_record
plt.bar(np.arange(bat_season_record.shape[0])- 0.2, bat_season_record['2015'],width=0.2)
plt.bar(np.arange(bat_season_record.shape[0]), bat_season_record['2016'],width=0.2)
plt.bar(np.arange(bat_season_record.shape[0])+ 0.2, bat_season_record['2017'],width=0.2)
plt.xticks(np.arange(bat_season_record.shape[0]), bat_season_record['batsman'])


plt.show()
```

```python
# 3d surface plot
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx ,yy = np.meshgrid(x,y)
```

```python

z=  xx**2 + yy**2
z.shape
```

```python

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python

z = np.sin(xx)+np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python

z = np.sin(xx)+np.log(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# contour plot 3d grpah to 2d mein krey ga
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# contour plot 3d grpah to 2d mein krey ga
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
