```python
# what are vectorized operations?
# vectorized operations are operations that are applied to entire arrays or series at once, rather than element by element.
# this can lead to more efficient and readable code, as well as better performance in many cases.
# for example, consider the following code that adds two arrays element by element:
s = pd.Series(['cat','mat',None,'rat'])
s.str.startswith('c')
```

```python
df = pd.read_csv('dataset/titanic.csv')
```

```python
df['Name'].str.lower()
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()
# len
df['Name'][df['Name'].str.len() == 82].values[0]
#strip
df['Name'].str.strip()
```

```python
#split()
df["LastName"] = df["Name"].str.split(',').str.get(0)
df[['title','FristName']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n = 1,expand=True)
```

```python
df['title'].value_counts()
```

```python
#replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
# filtereing
# startwith and endwith
df['FristName'].str.startswith('A')
# isdigit / isalpha
df['FristName'].str.isalpha()
```

```python
# regex applying
# search john in both case
df[df['FristName'].str.contains('john',case=False)]

# find the lkastanme with start and end with vowels
df[df['FristName'].str.contains('^[^aeiouAEIUO].+[aeiouAEIUO]$')]
```
