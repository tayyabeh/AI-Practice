```python
# left join (comon and complete left table)
course.merge(reg,how = 'left', on = 'course_id') # left join
```

```python
# right join (comon and complete right table)
temp = pd.DataFrame({
    'student_id': [21, 32, 23, 24],
    'name': ['A', 'B', 'C', 'D'],
    'partner': [26, 27, 34, 45]
})
students = pd.concat([students, temp],ignore_index=True) # vertical concatenation
```

```python
# 1 . find the total revenue generated by the company
tot_rev = reg.merge(course, how = 'inner', on = 'course_id')['price'].sum()
print("total revenue generated by the company is : ", tot_rev)
```

```python
# 2. find the month by month revenue
temp = pd.concat([nov, dec],keys=['Nov','Dec']).reset_index()
temp.merge(course, how = 'inner', on = 'course_id').groupby('level_0')['price'].sum() # month by month revenue
```

```python
# 3, print the registration table
# cols -> name -> price
reg.merge(students, on = "student_id").merge(course, on = 'course_id')[['name','course_name','price']]
```

```python
# 4. plot bar chart for revenue / course
reg.merge(course,on='course_id').groupby('course_name')['price'].sum().plot(kind='bar', figsize=(10,5), title='Revenue per course')
```

```python
# 5. find the students who enrolled in  the both the months
com_st_id = np.intersect1d(nov['student_id'],dec['student_id'])
students[students['student_id'].isin(com_st_id)]
```

```python
# print student name -> partner name for all enrolled studnets
# self join 
students.merge(students, how = 'inner', left_on='partner', right_on = 'student_id')[['name_x','name_y']]
```

```python
# find top 3 students who did most number enrollments
reg.merge(students, on ='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3) # top 3 students who did most number enrollments
```

```python
# find the top 3 students who spent most amount ofmoney on courses
reg.merge(students, on ='student_id').merge(course , on = 'course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3) # top 3 students who spent most amount ofmoney on courses
```

```python
# ipl problem 

# find top 3 stadiums with highest sixes / matches ratio
temp = deliveries.merge(matches, how = 'inner', left_on = 'match_id',right_on ='id')
```

```python
six_temp = temp[temp['batsman_runs'] == 6]
# sixes per stadium  ( stadium , sixes)
num_six = six_temp.groupby('venue')['venue'].count()
```

```python
num_matches = matches['venue'].value_counts()
```

```python
(num_six/num_matches).sort_values(ascending=False).head(3) # top 3 stadiums with highest sixes / matches ratio
```

```python
# find orange cap holder of every year
temp.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep = 'first').sort_values('season') # orange cap holder of every year
```
