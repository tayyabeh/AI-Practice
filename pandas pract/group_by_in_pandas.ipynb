```python
import pandas as pd 
import numpy as np 
```

```python
movies = pd.read_csv('dataset\imdb-top-1000.csv')
```

```python
movies.head()
```

```python
genres = movies.groupby('Genre')
```

```python
# APply bultin aggregate functions 
# mean, sum, count, min, max
genres.sum()
```

```python
# find the top 3 genre by total earning
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with the highest average rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(3)
```

```python
# find the director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(3)
```

```python
# find the higest rated movie of each genre
movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False).head(3)
```

```python
# find the no. of movies done by each actor
# movies['Star1'].value_counts()
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False).head(3)
```

```python
# Groupby attributes and methods
# find total no. of group
# len (movies.groupby('Genre'))

# how many rows in group 
# movies.groupby('Genre').size()

# first / last item of very group
genre = movies.groupby('Genre')
# genre.first()
# genre.last()
# genre.nth(6)

# get the group of each genre
genre.get_group('Horror')

# groups 
genre.groups # index of movies of every genre

# describe 
movies.groupby('Genre').describe()

#sample
genre.sample(2,replace = True)

# nunique
genre.nunique()
```

```python
# agg methods
# passing dict
# genre.agg({
#     'Runtime':'mean',
#     'IMDB_Rating':'mean',
#     'No_of_Votes':'sum',
#     'Gross':'sum',
#     'Metascore':'mean'
# })

# passing list 
genre.agg(['min','max','sum'])


# adding the both 

genre.agg({
    'Runtime': ['mean','max'],
    'IMDB_Rating':['mean','max'],
    'No_of_Votes':['sum','max'],
    'Gross':['sum','mean'],
    'Metascore':'mean'
})
```

```python
# # looping on groups
# df = pd.DataFrame(columns =movies.columns)
# for group,data in genre :
#     df = df.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
# # version per aab nhi hai 
```

```python
df_list = []

for group, data in genre:
    top_rated = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df_list.append(top_rated)

df = pd.concat(df_list, ignore_index=True)
df.head(17)
```

```python
# split( apply) combine 
# split - groupby
# apply - aggregate function
# combine - concat
# split - apply - combine

genre.apply(min)
```

```python
# find the numbre of movies staring with A for each group

def foo ( group):
    # print(group['Series_Title'].str.startswith('A').sum())
    return group['Series_Title'].str.startswith('A').sum()

genre.apply(foo)
```

```python
# find the ranking of each movie in thte group according to IMDB score
def rank(group):
    group['Rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group


genre.apply(rank).head(10)
```

```python
# find the  normalized IMDB rating group group wise

def normal(group):
    group['norm_rating '] = (group['IMDB_Rating'] - group['IMDB_Rating'].mean()) //( group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

genre.apply(normal)
```

```python
# groupby on multiple cols 
duo = movies.groupby(['Director','Star1'])
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# find the most earning actor --> director combo
duo['Gross'].sum().sort_values(ascending=False)
```

```python
# find the best actor--> genre combo in meta score wise
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values(by='Metascore', ascending=False).head(10)
```

```python
# agg on multilpe groupby
duo.agg(['min','max','sum'])
```

```python
# find the top 10 batsman in terms of runs scored 
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no. of sixes 
sixes = ipl[ipl['batsman_runs']== 6]

sixes.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(10)
```

```python
# find the batsman with most no. of 4's and 6's in lat 5 overs
last_overs = ipl[ipl ['over']> 15 ]
last_overs = last_overs[(last_overs['batsman_runs'] == 4) | (last_overs['batsman_runs'] == 6)]
last_overs.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(10)
```

```python
# Vriat kohli record against all teams
vk = ipl[ipl['batsman'] == 'V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the higest score of any batsman
vk = ipl[ipl['batsman'] == 'V Kohli']
vk.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
def highest_score(batsman):
    player = ipl[ipl['batsman'] == batsman]
    return player.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```
